```python
# Sigmoid function
import numpy as np
def sigmoid(z):
  return 1.0/(1 + np.exp(-np.dot(z, b)))


def predict(inputs):
  pred_prob = sigmoid(inputs)
  pred_value = np.where(pred_prob >= .5, 1, 0)
  return np.squeeze(pred_value)
```
